```python
popu_seoul = pd.read_csv(
    "./data/2017년부터 서울시 인구수.csv", header=2)
popu_seoul=popu_seoul[["자치구","계", "계.1", "계.2", "65세이상고령자"]]
popu_seoul
```

```python
popu_seoul.rename(
    columns={
        popu_seoul.columns[0]: "구별",
        popu_seoul.columns[1]: "인구수",
        popu_seoul.columns[2]: "한국인",
        popu_seoul.columns[3]: "외국인",
        popu_seoul.columns[4]: "고령자",
    },
    inplace=True
)
popu_seoul
```

```python
CCTV_Seoul["최근증가율"] = (
    (CCTV_Seoul["2021년"] +CCTV_Seoul["2020년"] +CCTV_Seoul["2019년"] +CCTV_Seoul["2018년"] +CCTV_Seoul["2017년"] +CCTV_Seoul["2016년"] + CCTV_Seoul["2015년"] + CCTV_Seoul["2014년"]) / CCTV_Seoul["2012년 이전"] * 100
)

CCTV_Seoul.sort_values(by="최근증가율", ascending=False)
```

```python
popu_seoul["구별"].unique()
```

```python
popu_seoul["외국인비율"] = popu_seoul["외국인"] / popu_seoul["인구수"] * 100
popu_seoul["고령자비율"] = popu_seoul["고령자"] / popu_seoul["인구수"] * 100
popu_seoul["최근증가율"]=CCTV_Seoul["최근증가율"]
popu_seoul["CCTV 설치비율"]=CCTV_Seoul["총계"]/833557 *100
popu_seoul["소계"]=CCTV_Seoul["총계"]
data_result=popu_seoul[:26]
data_result=data_result.drop([data_result.index[0]])
data_result.to_csv("서울 cctv.csv")
```
