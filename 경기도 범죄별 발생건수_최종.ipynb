```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
crime=pd.read_csv('./data/경기도 범죄별 범죄건수.csv', encoding='cp949')
crime
```

```python
#crime=crime.set_index('범죄중분류')
crime_df=crime.T
crime_df
```

```python
crime_df['계']=crime_df.sum(axis=1)
crime_df
```

```python
crime_df = crime_df.reset_index()
new_crime_df=crime_df[['index', '계']].drop(0)
new_crime_df
```

```python
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

le=LabelEncoder()
new_crime_df['index']=le.fit_transform(new_crime_df['index'])
new_crime_df
```

```python
y_target=new_crime_df['계']
X_ftr=new_crime_df.drop('계', axis=1)
```

```python
from sklearn.model_selection import  train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error
X_train, X_test, y_train, y_test=train_test_split(X_ftr, y_target)
```

```python
# 학습
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train,y_train)

y_pred = dt_reg.predict(X_test)
mae = mean_absolute_error(y_test,y_pred)
```

```python
check_df=pd.DataFrame()
check_df['actual_y']=y_test
check_df['pred_y']=y_pred
check_df['diff']=np.abs(check_df['actual_y']-check_df['pred_y'])

check_df.sort_values(by=['diff'], ascending=False)
```

```python
check_df['diff_percentage'] = check_df['diff']/check_df['actual_y'] * 100
check_df.sort_values(by=['diff_percentage'], ascending=False)[:10]
```
